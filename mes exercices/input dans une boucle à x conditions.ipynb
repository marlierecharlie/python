```python
# input : utiliser une boucle while plutôt que for

def time():
    day = True
    date = []
    while day : 
        d = input ("Entrez le jour d'arrivée : ")
        if d.isdigit() and len(d)==2 and 1 <= int(d) <=31 : 
#        1 <= int(d) <=31  permet de ne rentrer que des valeurs comprises entre 1 et 31     
            d=int(d)
#      d=int(d) permet de tranformer des caractères en nombres
            day = False 
            date.append(d)
        else :
            print ("entrez deux chiffres")         
    return date

time()
```
